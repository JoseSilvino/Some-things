```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgs = []
for i in range(1,4):
  plt.subplot(1,3,i)
  imgs.append(cv2.imread(f"Img{i}.png"))
  imgs[i-1] = cv2.cvtColor(imgs[i-1],cv2.COLOR_RGB2BGR)
  plt.imshow(imgs[i-1])
```

```python
stitcher = cv2.Stitcher_create()
(status,stitched) = stitcher.stitch(imgs)
if status == 0:
  plt.imshow(stitched)
else:
  print("[INFO] image stitching failed ({})".format(status))
```
