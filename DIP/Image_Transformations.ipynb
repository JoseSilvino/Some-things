```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
def readopenCV(src):
  img = cv2.imread(src)
  img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
  return img
```

```python
#Original image
image = Image.open("AguEGao.jpg")
plt.imshow(image)
```

```python
#Crop flip an image using Numpy array indexing.
image = Image.open("AguEGao.jpg")
image_arr = np.array(image)
image_arr = image_arr[400:600,300:500]
image = Image.fromarray(image_arr)
plt.imshow(image)
```

```python
#Flip an image using Numpy array indexing.
image = Image.open("AguEGao.jpg")
image_arr = np.array(image)
image_arr = np.flip(image_arr,1)
image = Image.fromarray(image_arr)
plt.imshow(image)
```

```python
#Implement image translation using Numpy and OpenCV.
image = readopenCV("AguEGao.jpg")
h,w = image.shape[:2]
plt.imshow(image)
qh,qw = h/5,w/5
T = np.float32([[1,0,qw],[0,1,qh]])
img_translation = cv2.warpAffine(image,T,(w,h))
plt.imshow(img_translation)
```

```python
#Implement image rotation using Numpy and OpenCV.
image = readopenCV("AguEGao.jpg")
image = cv2.rotate(image,cv2.cv2.ROTATE_90_CLOCKWISE)
plt.imshow(image)
```

```python
#Implement image resizing using OpenCV.
image = readopenCV("AguEGao.jpg")
half = cv2.resize(image,(0,0),fx = 0.5,fy = 0.5)
bigger = cv2.resize(image,(h*2,w*2))
stretch = cv2.resize(image,(780,540),interpolation=cv2.INTER_NEAREST)
titles = ["Original","half","Bigger","Interpolation nearest"]
images = [image,half,bigger,stretch]
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.title(titles[i])
  plt.imshow(images[i])
plt.show()
```

```python
#Implement bitwise operations: AND, OR, XOR.
img1 = readopenCV("AguEGao.jpg")
img2 = readopenCV("Gatch&Musi&Doka.jpg")
andimg = cv2.bitwise_and(img1,img2,mask = None)
orimg = cv2.bitwise_or(img1,img2,mask = None)
xorimg = cv2.bitwise_xor(img1,img2,mask = None)
images = [img1,img2,andimg,orimg,xorimg]
titles = ["Original1","Original2","And","Or","Xor"]
for i in range(5):
  plt.subplot(2,3,i+1)
  plt.title(titles[i])
  plt.imshow(images[i])
```

```python
def getcircle():
  circle_arr = np.zeros(shape=(600,700),dtype= np.uint8)
  center = (350,300)
  radius = 200
  color = 255
  img2 = cv2.circle(circle_arr,center,radius,color,-1)
  return img2
```

```python
#Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.
img1 = readopenCV("AguEGao.jpg")
img2 = getcircle()
#cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
plt.subplot(2,1,1)
plt.title("IMG1")
plt.imshow(img1)
plt.subplot(2,1,2)
plt.title("IMG2")
plt.imshow(img2)
```
